```python
# PANDAS 패키지 불러오기
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
studentData = pd.read_csv("../dataset/student_middle_result.csv")

studentData.head()
```

```python
name = studentData[ ['ST_NO']]
feature = studentData[ ['SW','DB','AND']]
feature.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
#import seaborn as sns

# create model and prediction
model = KMeans(n_clusters=2,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']

# concatenate labels to df as a new column
r = pd.concat([name,feature,predict],axis=1)

r.head()
```

```python
# CSV 파일로 저장
r.to_csv("../dataset/student_out.csv", encoding='utf-8-sig', index=False)
r.head()
```

```python
# scatter plot
plt.scatter(r['DB'], r['AND'],c=r['PREDICT'],alpha=0.5)
#plt.scatter(center_x,center_y,s=50,marker='D',c='r')
plt.show()
```

```python
# PANDAS 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
# CSV 파일을 읽어 Data Frame 변수에 저장하기
studentData = pd.read_csv("../dataset/student_middle_result.csv")

name = studentData[ ['ST_NO']]
feature = studentData[ ['SW','DB','AND']]

# create model and prediction
model = KMeans(n_clusters=2,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']

# concatenate labels to df as a new column
r = pd.concat([name,feature,predict],axis=1)

# scatter plot
plt.scatter(r['DB'], r['AND'],c=r['PREDICT'],alpha=0.5)
#plt.scatter(center_x,center_y,s=50,marker='D',c='r')
plt.show()
```
