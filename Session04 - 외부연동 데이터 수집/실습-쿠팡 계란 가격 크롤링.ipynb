```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
from os.path import join, dirname
```

```python
import requests, bs4 
import pandas as pd
```

```python
# 크롬 옵션 설정
options = webdriver.ChromeOptions()
# 헤드리스 설정 시 리눅스 같이 웹화면 없는경우에도 웹페이지 실행가능함
#options.add_argument('headless')
options.add_argument('window-size=1920x1080')
#options.add_argument("disable-gpu")
# 웹 드라이버 정의
driver = webdriver.Chrome(dirver_loc, options=options)
```

```python
from selenium.webdriver.common.action_chains import ActionChains
coupangUrl = 'http://www.coupang.com'
driver.get(coupangUrl)

# 카테고리 메뉴 (카테고리 -> 채소 )
mainMenu = '//*[@id="header"]/div/a'
subMenu = '//*[@id="gnbAnalytics"]/ul[1]/li[4]/a'

main = driver.find_element_by_xpath(mainMenu)
sub = driver.find_element_by_xpath(subMenu)

# 메인 이동 후 클릭
ActionChains(driver).move_to_element(main).click(sub).perform()
```

```python
emptyList =[]
for i in range(0, 99999):
    print(i)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',}
    resp=requests.get("https://www.coupang.com/np/categories/194810?page="+str(i+1),headers=headers)
    
        
    resp.encoding='utf-8' 
    html = resp.text 

    bs = bs4.BeautifulSoup(html, 'html.parser') 

    if len(bs.find_all("p",{"class":"no-list-item"})) > 0:
        break
        
    # html 태그 부분구조 가져오기 
    #find_tag = bs.find_all("div") 
    find_attr = bs.find("ul",{"class":"baby-product-list"}) 

    nameList = []
    priceList = []
    rowList = []
    s_li = find_attr.find_all("li")
    s_div = s_li[0].find_all("div")
    s_div[1].text.replace("\n","")

    for i in range(0,len(s_li)):
        s_div = s_li[i].find_all("div")
        rowList.append(s_div[1].text.replace("\n",""))
        priceList.append(s_div[2].find("strong").text)

    bsout = pd.DataFrame(zip(rowList,priceList))

    bsout.columns = ["제품명","가격"]
    bsout["페이지"] = i+1
    emptyList.append(bsout)
bsout
```
