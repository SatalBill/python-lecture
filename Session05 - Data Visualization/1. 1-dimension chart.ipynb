### 2차원 차트 그리기

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시여
%matplotlib inline

x=list(range(0,100,5))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(x,y, 'b', lw=1.5, label = '1st')
plt.plot(x,y, 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx()
plt.plot(x,y2, 'g', lw=1.5, label = '2nd')
plt.plot(x,y2, 'go')
plt.legend(loc=2)
plt.ylabel('value 2nd')
```

### 2차원 Subplot

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시여
%matplotlib inline

x=list(range(0,100,5))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```
